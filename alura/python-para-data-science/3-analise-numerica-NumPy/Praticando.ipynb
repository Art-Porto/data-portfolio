```python
import numpy as np
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/refs/heads/dados/citrus.csv'
dado = np.loadtxt(url, delimiter=',', skiprows=1, usecols=np.arange(1,3,1))
diametro_laranja = dado[:5000,0]
diametro_toranja = dado[5000:,0]
peso_laranja = dado[:5000,1]
peso_toranja = dado[5000:,1]

import matplotlib.pyplot as plt

plt.scatter(diametro_laranja, peso_laranja, marker= 'x')
plt.scatter(diametro_toranja, peso_toranja, marker= 'x')
```

```python
import numpy as np

a_laranja, b_laranja = np.polyfit(peso_laranja, diametro_laranja,1)
a_toranja, b_toranja = np.polyfit(peso_toranja, diametro_toranja, 1)
y_laranja = a_laranja * peso_laranja + b_laranja
y_toranja = a_toranja * peso_toranja + b_toranja
```

```python
import numpy as np
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/refs/heads/dados/citrus.csv'
dado = np.loadtxt(url, delimiter=',', skiprows=1, usecols=np.arange(1,3,1))
diametro_laranja = dado[:5000,0]
diametro_toranja = dado[5000:,0]
peso_laranja = dado[:5000,1]
peso_toranja = dado[5000:,1]

import matplotlib.pyplot as plt

plt.scatter(diametro_laranja, peso_laranja, marker= 'x', s=0.5)
plt.scatter(diametro_toranja, peso_toranja, marker= 'x', s=0.5)
plt.plot(y_laranja, peso_laranja, c='r')
plt.plot(y_toranja, peso_toranja, c='r')
```
